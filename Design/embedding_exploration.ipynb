## **Method 1 for Sentence Embedding :** embedding a sentence by averaging the embeddings of its words

```python
sentence = "Hello, have you done the research for tomorrow's meeting?"
L = sentence.split()

print(L)

def embed_sentence(L):
    sentence_embedding = np.zeros(300)
    for word in L:
        # remove punctuation
        word = word.strip('.,?!"\'').lower()
        try:
            vec = wv[word]
            sentence_embedding += vec
        except KeyError:
            print(word, 'not in vocabulary, skipping...')
            continue
    sentence_embedding /= len(L)
    return sentence_embedding

embed_sentence(L)
```
